```python
import numpy as np
import importlib
import ag2025 as ag
importlib.reload(ag)
```

```python
dataset = 'Uji'
file_name = f"{dataset}.npz"
theta = 1
k_value = 3
testing_rss,testing_coordinates,training_rss,training_coordinates = ag.get_data(dataset,fill_invalid_with = -100)
normalized_rss = ag.rss_normalize(training_rss,'z')
```

```python
data = np.load(file_name)
```

```python
Pred_list = np.zeros_like(testing_coordinates)

# Iterate through test samples
for i, row in enumerate(testing_rss):
    train_X, train_Y = ag.getfp(training_rss, training_coordinates, normalized_rss, row, cos_p=theta)
    KNN_reg = ag.KNNRegression(k=k_value)
    KNN_reg.fit(train_X, train_Y)
    y_pred = KNN_reg.predict(row.reshape(1, -1))
    Pred_list[i, :] = y_pred
```

```python
tsfa_error = ag.euclidean_distance(Pred_list[:,-3:], testing_coordinates[:,-3:])
```

```python
m_t = np.mean(tsfa_error)
var_t = np.var(tsfa_error,ddof = 1)

print(f'GMM-TSFA - Mean: {m_t:.2f}, Variance: {var_t:.2f}')
```
