```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset_name = 'TUT1'
sknn_path =  'results/'+dataset_name+'_n100_results.csv'
knn_path = 'results/'+dataset_name+'_knn.csv'
```

```python
rs_sknn = pd.read_csv(sknn_path)
rs_knn = pd.read_csv(knn_path)
```

```python
knn_mae = rs_knn[['k','MAE']].values
sknn_mae = rs_sknn[['cos_p','k','MAE']].values
```

```python
pct_v = np.asarray(np.unique(sknn_mae[:,0]))
```

```python
plt.figure()

for i in pct_v:
    data_to_plot = sknn_mae[sknn_mae[:,0]==i,-2:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], marker='o', linestyle='-',label=f'theta: {i/100}')

plt.plot(knn_mae[:,0], knn_mae[:,1], marker='>', linestyle='-', label='KNN',color = 'black')
# ttl = dataset_name + ': k vs mae' + method
ttl = dataset_name + ': k vs MAE'
plt.title(ttl)
plt.xlabel('k values')
plt.ylabel('MAE(m)')
plt.grid(True)
plt.figlegend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

```python
knn_80p = rs_knn[['k','P80']].values
sknn_80p = rs_sknn[['cos_p','k','P80']].values
```

```python
plt.figure()

for i in pct_v:
    data_to_plot = sknn_80p[sknn_80p[:,0]==i,-2:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], marker='o', linestyle='-',label=f'theta: {i/100} ')

plt.plot(knn_80p[:,0], knn_80p[:,1], marker='>', linestyle='-', label='KNN',color = 'black')


# ttl = dataset_name + ': k vs 80% error' + method
ttl = dataset_name + ': k vs 80% error'
plt.title(ttl)
plt.xlabel('k values')
plt.ylabel('80 Percentile Error(m)')
plt.grid(True)
plt.figlegend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```
